```python
from pinecone import Pinecone, ServerlessSpec

pc = Pinecone(api_key="")
```

```python
# Creating Langchain index (dense Index)

from pinecone import ServerlessSpec

index_name = "langchain-test-index"

if not pc.has_index(index_name):
    pc.create_index(
        name=index_name,
        dimension=3072,
        metric="cosine",
        spec=ServerlessSpec(cloud="aws", region="us-east-1"),
    )

index = pc.Index(index_name)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

GOOGLE_API_KEY = ""

embedding_function = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001",google_api_key=GOOGLE_API_KEY)

```

```python
from langchain_pinecone import PineconeVectorStore

vector_store = PineconeVectorStore(index=index, embedding=embedding_function)
```

```python
from langchain.schema import Document

doc1 = Document(
        page_content=("Virat Kohli is one of the most consistent batsmen in IPL history. He has led the Royal Challengers Bengaluru with passion and intensity."),
        metadata={"team": "Royal Challengers Bengaluru"}
      )

doc2 = Document(
        page_content=("MS Dhoni is a legendary finisher and one of the best captains in IPL.He has guided Chennai Super Kings to multiple IPL titles."),
        metadata={"team": "Chennai Super Kings"}
      )

doc3 = Document(
        page_content=("Ravindra Jadeja is a dynamic all-rounder known for his sharp fielding. He plays a crucial role for Chennai Super Kings in both batting and bowling."),
        metadata={"team": "Chennai Super Kings"}
      )

doc4 = Document(
        page_content=("Jasprit Bumrah is one of the most lethal fast bowlers in IPL. He has been a key match-winner for Mumbai Indians with his death bowling."),
        metadata={"team": "Mumbai Indians"}
      )

doc5 = Document(
        page_content=("Rohit Sharma is known for his calm leadership and explosive batting. He has captained Mumbai Indians to multiple championship victories."),
        metadata={"team": "Mumbai Indians"}
     )
```

```python
docs = [doc1,doc2,doc3,doc4,doc5]
```

```python
vector_store.add_documents(docs)
```

```python
# Get/View documents from Chrom DB

index = pc.Index(host="https://langchain-test-index-17534yg.svc.aped-4627-b74a.pinecone.io")

index.fetch(ids=["ff06250f-c3ba-4be4-88ae-57d74201bc13"], namespace="__default__")
```

```python
# search documents

vector_store.similarity_search(
    query="Who is bowler",
    k=2
)
```

```python
# search documents

vector_store.similarity_search(
    query="Who is the captain of Mumbai Indians ?",
    k=1
)
```

```python
# search documents based on the similarity score.
# if distance is close to 0 it represent a very high similarity between query and that document that is retreived.
# strive for less similarity score.

vector_store.similarity_search_with_score(
    query="Virat Kholi is the Captain of which IPL Team ?",
    k=2
)
```

```python
# We can filter data not only on the basis of Documents but also on the basis of metadata
# Metadata Filtering

# search documents

vector_store.similarity_search_with_score(
    query="player",
    filter={"team":"Chennai Super Kings"}
)
```

```python
# Fetching ID of any specific document

vector_store.similarity_search_with_score(
    query = "Virat",
    k=1
)
```

```python
# Update both Document and Metadata

new_vector = embedding_function.embed_query("Virat Kholi")  # numeric embedding
index.update(
    namespace="__default__",
    id="ff06250f-c3ba-4be4-88ae-57d74201bc13",
    values=new_vector,       # must be list of floats
    set_metadata={"team": "RCB"}
)
```

```python
# Update Metadata only

index.update(
    namespace="__default__",
    id="ff06250f-c3ba-4be4-88ae-57d74201bc13",
    set_metadata={"team": "Royal Bangalore"}
)
```

```python
# delete records based on ID

index.delete(
    ids=["ff06250f-c3ba-4be4-88ae-57d74201bc13"],
    namespace="__default__"
)
```
